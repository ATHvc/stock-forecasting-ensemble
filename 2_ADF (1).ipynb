```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
```

```python
df = pd.read_csv(
    "/content/pltr_ts_processed.csv",
    index_col=0,
    parse_dates=True
)

df.head()
```

```python
ts_return = df['log_return'].dropna()
```

```python
adf_result = adfuller(ts_return)
```

```python
labels = [
    "ADF Statistic",
    "p-value",
    "# Lags Used",
    "# Observations Used"
]

for label, value in zip(labels, adf_result[:4]):
    print(f"{label}: {value}")
```

```python
print("\nCritical Values:")
for key, value in adf_result[4].items():
    print(f"{key}: {value}")
```
