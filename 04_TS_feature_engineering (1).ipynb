```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "/content/pltr_ts_processed.csv",
    index_col=0,
    parse_dates=True
)

df.head()
```

```python
ret = df['log_return']
```

```python
for lag in [1, 2, 5, 10]:
    df[f'lag_{lag}'] = ret.shift(lag)
```

```python
df['roll_mean_5']  = ret.rolling(5).mean()
df['roll_std_5']   = ret.rolling(5).std()

df['roll_mean_20'] = ret.rolling(20).mean()
df['roll_std_20']  = ret.rolling(20).std()
```

```python
df['ewma_10'] = ret.ewm(span=10).mean()
df['ewma_20'] = ret.ewm(span=20).mean()
```

```python
df['abs_return'] = ret.abs()
df['sq_return']  = ret ** 2
```

```python
df['target'] = ret.shift(-1)
```

```python
df_fe = df.dropna()
```

```python
feature_cols = [
    'lag_1','lag_2','lag_5','lag_10',
    'roll_mean_5','roll_std_5',
    'roll_mean_20','roll_std_20',
    'ewma_10','ewma_20',
    'abs_return','sq_return'
]

X = df_fe[feature_cols]
y = df_fe['target']
```

```python
X.to_csv("/content/X_features.csv")
y.to_csv("/content/y_target.csv")
```
